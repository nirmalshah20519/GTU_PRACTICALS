### Q13. Write a Pandas program to drop all non-numeric columns from diamonds DataFrame.

```python
array=[]
for _ in diamond1.columns:
    if diamond1[_].dtype=='object':
        array.append(_)
diamond1.drop(columns=array,inplace=True)
diamond1
```
